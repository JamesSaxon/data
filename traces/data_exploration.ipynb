```python
df = pd.read_sql("SELECT DATE_TRUNC('HOUR', time) date, AVG((prec = 0)::integer) "
                 "FROM traces GROUP BY DATE_TRUNC('HOUR', time);",
                 con = tr_con, parse_dates = ["date"], index_col = "date")

ax = df.avg.plot()
ax.set_xlabel("")
ax.set_ylabel("Fraction Imprecise")

ax.figure.savefig("figs/timing_imprecise.pdf")
```
